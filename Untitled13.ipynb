```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt

# Define paths to your datasets (original artworks and new image)
original_dataset_path ='/content/drive/MyDrive/ABBOTTS BABBLER'
new_image_path = '/content/drive/MyDrive/ABBOTTS BABBLER/005.jpg'

# Load and preprocess images
def load_and_preprocess_image(image_path, target_size=(224, 224)):
    img = load_img(image_path, target_size=target_size)
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    return img_array

# Create a VGG16 model for feature extraction
def create_vgg16_model():
    base_model = VGG16(weights='imagenet', include_top=False)
    base_model.trainable = False

    # Choose intermediate layers for style and content representation
    style_layers = ['block1_conv2', 'block2_conv2', 'block3_conv3', 'block4_conv3']
    content_layer = 'block5_conv3'

    style_outputs = [base_model.get_layer(layer).output for layer in style_layers]
    content_output = base_model.get_layer(content_layer).output

    model = Model(inputs=base_model.input, outputs=style_outputs + [content_output])
    return model

# Style transfer function
# Function to perform style transfer with masking
# Simplified style transfer function with comments
def style_transfer_simple(base_img, style_img, content_img, iterations=100, content_weight=1e3, style_weight=1e-2, tv_weight=2e2):
    # Create the VGG16 model
    vgg16_model = create_vgg16_model()

    # Define the generated image as a trainable variable
    generated_img = tf.Variable(content_img, dtype=tf.float32)

    # Define the optimizer
    optimizer = Adam(learning_rate=5, beta_1=0.99, epsilon=1e-1)

    # Training loop
    for i in range(iterations):
        with tf.GradientTape() as tape:
            model_outputs = vgg16_model(generated_img)
            style_outputs = model_outputs[:-1]
            content_output = model_outputs[-1]

            # Calculate content loss
            content_loss = content_weight * tf.reduce_mean(tf.square(content_output - vgg16_model(base_img)[-1]))

            # Calculate style loss
            style_loss = 0
            for target, gen in zip(style_outputs, vgg16_model(generated_img)[:-1]):
                style_loss += style_weight * tf.reduce_sum(tf.square(target - gen))

            # Calculate total variation loss to reduce noise
            tv_loss = tv_weight * tf.reduce_sum(tf.image.total_variation(generated_img))

            # Total loss is the sum of content, style, and total variation losses
            total_loss = content_loss + style_loss + tv_loss

        # Calculate gradients and apply to the generated image
        gradients = tape.gradient(total_loss, generated_img)
        optimizer.apply_gradients([(gradients, generated_img)])
        generated_img.assign(tf.clip_by_value(generated_img, clip_value_min=0.0, clip_value_max=255.0))

        if i % 10 == 0:
            print(f"Iteration {i}/{iterations}, Total Loss: {total_loss.numpy()}")

    # Display the final stylized image
    result_img = generated_img.numpy().squeeze()
    plt.imshow(result_img.astype('uint8'))
    plt.axis('off')
    plt.show()

    return result_img


# Load a new image for style adaptation
new_image_path = '/content/drive/MyDrive/ABBOTTS BABBLER/005.jpg'
new_image = load_and_preprocess_image(new_image_path)

# Apply style transfer to generate the learned style
learned_style = style_transfer(original_image, original_image, iterations=200)

# Load a new image for style adaptation
new_image = load_and_preprocess_image(new_image_path)

# Apply the learned style to the new image
stylized_image = style_transfer(new_image, learned_style, iterations=100)

# Display the original, learned style, and stylized images
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(original_image.squeeze().astype('uint8'))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(learned_style.astype('uint8'))
plt.title('Learned Style')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(stylized_image.astype('uint8'))
plt.title('Stylized Image')
plt.axis('off')

plt.show()

```
